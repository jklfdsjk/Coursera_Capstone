```python
def getNearbyVenues(names, latitudes, longitudes, radius=7000):
    LIMIT = 100
    venues_list=[]
    for name, la, ln in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            la, 
            ln, 
            VERSION, 
            search_query, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()
        
        # return only relevant information for each nearby venue
        venues_list.append([(
             name, 
             la, 
             ln, 
             i['name']) for i in results['response']['venues']])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['City', 
                  'City Latitude', 
                  'City Longitude', 
                  'Hospital']
    
    return(nearby_venues)   
```

```python
Cali_hospital = getNearbyVenues(names=df['City'],
                                   latitudes=df['Lat'],
                                   longitudes=df['Lng']
                                  )
```

```python
Cali_hospital.shape
Cali_hospital_drop = Cali_hospital.drop_duplicates(subset = 'Hospital', keep ='first')
print(Cali_hospital_drop.shape)
Cali_hospital_drop.tail()


Cali_hospital_drop = Cali_hospital_drop[Cali_hospital_drop['Hospital'].str.contains('Pet Clinic')]
Cali_hospital_drop
```

```python
count = Cali_hospital_drop.groupby(['City Latitude', 'City Longitude']).size().reset_index(name='counts')
count.counts.sum()
```

```python
df_merged = df.join(count.set_index(['City Latitude', 'City Longitude']), on= ['Lat', 'Lng'])
df_merged = df_merged.dropna()
df_merged = df_merged.sort_values("counts", ascending = False)
df_merged

df_merged.to_csv('Pet_Clinic.csv')
```
