```python
import requests
import pandas as pd 
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df = pd.read_html(url)
df = df[0]
df.columns = df.iloc[0]
df = df.drop(df.index[0])
df = df[df.Borough != 'Not assigned']
df
```
