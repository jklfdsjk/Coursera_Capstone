```python
import numpy as np 
from sklearn.cluster import KMeans 
from sklearn.preprocessing import StandardScaler 
```

```python
X = df[['Population', 'Income', 'median_house_value']].values 
X = np.nan_to_num(X)
cluster_dataset = StandardScaler().fit_transform(X)
cluster_dataset
```

```python
K = range (1, 15)

best_k = []
for k in K:
    kmeans_model = KMeans(n_clusters=k, random_state=2)
    kmeans_model = kmeans_model.fit(cluster_dataset) # Create a kmeans model on our data, using k clusters.  random_state helps ensure that the algorithm returns the same results each time.
    labels = kmeans_model.labels_ # These are our fitted labels for clusters -- the first cluster has label 0, and the second has label 1.
    inertia = kmeans_model.inertia_ # Sum of distances of samples to their closest cluster center
    best_k.append(inertia)
   
    
best_k
```

```python
plt.plot(K, best_k, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
num_clusters = 4

k_means = KMeans(init="k-means++", n_clusters=num_clusters, n_init=12)
k_means.fit(cluster_dataset)
labels = k_means.labels_

print(labels)

np.unique(labels)
```

```python
def getNearbyVenues(names, latitudes, longitudes, search_query, radius=4000):
    LIMIT = 100
    
    venues_list=[]
    for name, la, ln in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            la, 
            ln, 
            VERSION, 
            search_query, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()
        
        # return only relevant information for each nearby venue
        venues_list.append([(
             name, 
             la, 
             ln, 
             i['name']) for i in results['response']['venues']])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['City', 
                  'City Latitude', 
                  'City Longitude', 
                  'Hospital']
    
    return(nearby_venues)   
```
